```python
import seaborn as sns
import pandas as pd

url = 'https://raw.githubusercontent.com/AnjulaMehto/MCA/main/company_sales_data.csv'
df = pd.read_csv(url)

df
```

```python
import matplotlib.pyplot as plt

monthList  = df['month_number'].tolist()
faceCremSalesData   = df ['facecream'].tolist()
faceWashSalesData   = df ['facewash'].tolist()

plt.bar([a-0.25 for a in monthList], faceCremSalesData, width= 0.25)
plt.bar([a+0.25 for a in monthList], faceWashSalesData, width= -0.25)


plt.xticks(monthList)
plt.xlabel('month list')
plt.ylabel('total sales')
plt.show()
```

```python
profitList = df ['total_profit'].tolist()
profit_range = [150000, 175000, 200000, 225000, 250000, 300000, 350000]
plt.hist(profitList, profit_range, label = 'Profit data')


plt.show()
```
